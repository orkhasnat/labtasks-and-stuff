```python
import pandas as pd
dse = pd.DataFrame()
for year in range(2008,2022):
  dse = pd.concat([dse,pd.read_json(f"prices_{year}.json")])
dse
```

```python
company = 'GRAMEENS2'
db = dse[dse['trading_code'] == company]
```

```python
from datetime import datetime
db['date'] = pd.to_datetime(db['date'])
db = db.set_index('date')
db = db.sort_index()
db.drop(columns=['trading_code'],axis=1,inplace=True)
db
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers import LSTM, Dropout, Dense
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

```python
db.info()
```

```python
scaler = StandardScaler()
scaler = scaler.fit(db)
scaled_data = scaler.transform(db)
scaled_data = scaled_data[:,~np.isnan(scaled_data).any(axis=0)]
scaled_data
```

```python
n_future = 1
n_past = 60
trainX=[]
trainY=[]
for i in range(n_past, len(scaled_data) - n_future + 1):
  trainX.append(scaled_data[i - n_past:i, 0:db.shape[1]])
  trainY.append(scaled_data[i + n_future - 1:i + n_future, 4])

trainX = np.array(trainX)
trainY = np.array(trainY)

print('trainX shape = {}'.format(trainX.shape))
print('trainY shape = {}'.format(trainY.shape))
```

```python
model = Sequential()
model.add(LSTM(128, activation='relu', input_shape=(trainX.shape[1],trainX.shape[2]), return_sequences=True))
model.add(LSTM(64, activation='relu', return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(trainY.shape[1]))
model.compile(optimizer='adam',loss='mse')
model.summary()
```

```python
history = model.fit(trainX, trainY, epochs=10, batch_size=16, validation_split=0.2, verbose=1)
```

```python
plt.plot(history.history['loss'], label='Training loss')
plt.plot(history.history['val_loss'], label='Validation loss')
plt.legend()
plt.show()
```

```python
n_future=60
forecast_period_dates=pd.date_range(db.index[-1], periods=n_future, freq='1d').tolist()
forecast = model.predict(trainX[-n_future:])
forecast_copies = np.repeat(forecast, db.shape[1],axis=-1)
y_pred_future = scaler.inverse_transform(forecast_copies)[:,0]
forecast_dates=[]
for time_i in forecast_period_dates:
  forecast_dates.append(time_i.date())

forecast = pd.DataFrame({'date':np.array(forecast_dates), 'opening_price':y_pred_future})
forecast['date'] = pd.to_datetime(forecast['date'])

actual = pd.read_json('prices_2022.json')
actual = actual[actual['trading_code'] == company]
actual['date'] = pd.to_datetime(actual['date'])
actual = actual.set_index('date')
actual = actual.sort_index()


plt.plot(db.index[-n_future:],db.opening_price[-n_future:],label='past')
plt.plot(forecast.date,forecast.opening_price,label='forecast')
plt.plot(actual.index[:n_future],actual.closing_price[:n_future],label='actual')
plt.legend()
plt.show()
```
